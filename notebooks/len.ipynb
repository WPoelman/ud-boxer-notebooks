```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style="darkgrid")
sns.despine()

DPI = 200

lang_mapping = {
    "en": "English",
    "de": "German",
    "nl": "Dutch",
    "it": "Italian",
}

placement = {
    0: (0),
    1: (1),
    2: (2),
    3: (3),
}

palette = {
    "U - Stanza": "#50b131",
    "U - Trankit": "#f77189",
    "N - Gold": "#bb9832",
    "N - Gold & Silver": "#464d49",
    "N - Gold (strict)": "#3ba3ec",
    "N - Gold & Silver (strict)": "#e866f4",
}

fig, axes = plt.subplots(4, 1, figsize=(10, 12), dpi=700)

for i, lang in enumerate(["de", "nl", "it", "en"]):
    df = pd.read_csv(f"analysis/combined_csv/{lang}_tok.csv")
    df["char_bins"] = pd.cut(
        df["n_char"],
        [1, 20, 40, 60, 80, 100, 120, 300],
        labels=[
            "1-20",
            "21-40",
            "41-60",
            "61-80",
            "81-100",
            "101-120",
            "120+",
        ],
    )
    ax = sns.lineplot(
        x="char_bins",
        y="value",
        hue="ud_system",
        ci=None,
        data=pd.melt(
            df[["f1", "char_bins", "ud_system"]],
            ["char_bins", "ud_system"],
            var_name="Scores",
        ),
        palette=palette,
        linewidth=1.0,
        ax=axes[placement[i]],
        legend=i == 3,
    )
    ax.set(ylim=(0, 1.0), xlim=(0, 6), xlabel="", ylabel="")
    # if i != 3:
    # ax.tick_params(bottom=False)
    # ax.set(xticklabels=[])

    if i == 0:
        ax.set(
            xticklabels=[
                "1-20\n" + r"$\bf{235}$",
                "21-40\n" + r"$\bf{722}$",
                "41-60\n" + r"$\bf{140}$",
                "61-80\n" + r"$\bf{8}$",
                "81-100\n" + r"$\bf{0}$",
                "101-120\n" + r"$\bf{0}$",
                "120+\n" + r"$\bf{0}$",
            ],
        )
    elif i == 1:
        ax.set(
            xticklabels=[
                "1-20\n" + r"$\bf{189}$",
                "21-40\n" + r"$\bf{615}$",
                "41-60\n" + r"$\bf{112}$",
                "61-80\n" + r"$\bf{11}$",
                "81-100\n" + r"$\bf{1}$",
                "101-120\n" + r"$\bf{0}$",
                "120+\n" + r"$\bf{0}$",
            ],
        )
    elif i == 2:
        ax.set(
            xticklabels=[
                "1-20\n" + r"$\bf{331}$",
                "21-40\n" + r"$\bf{601}$",
                "41-60\n" + r"$\bf{67}$",
                "61-80\n" + r"$\bf{2}$",
                "81-100\n" + r"$\bf{0}$",
                "101-120\n" + r"$\bf{0}$",
                "120+\n" + r"$\bf{0}$",
            ],
        )
    elif i == 3:
        ax.set(
            xlabel="Number of characters\n" + r"$\bf{Number\ of\ documents}$",
            xticklabels=[
                "1-20\n" + r"$\bf{756}$",
                "21-40\n" + r"$\bf{1880}$",
                "41-60\n" + r"$\bf{356}$",
                "61-80\n" + r"$\bf{33}$",
                "81-100\n" + r"$\bf{2}$",
                "101-120\n" + r"$\bf{4}$",
                "120+\n" + r"$\bf{3}$",
            ],
        )

        # Put a legend below current axis
        ax.legend(
            loc="upper center",
            bbox_to_anchor=(0.5, -0.45),
            ncol=3,
            facecolor="white",
        )
        # , fontsize="10",

    ax.set(ylabel=lang_mapping[lang])
    ax.set_ylabel(ax.get_ylabel(), rotation=0, horizontalalignment="right")

fig.supylabel("")
fig.suptitle("")
fig.tight_layout()


fig.savefig(f"analysis/scores_vs_len_char_interval.pdf")
```
