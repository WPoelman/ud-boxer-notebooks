```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style="darkgrid")
sns.despine()

DPI = 200

lang_mapping = {
    "en": "English",
    "de": "German",
    "nl": "Dutch",
    "it": "Italian",
}

placement = {
    0: (0),
    1: (1),
    2: (2),
    3: (3),
}

palette = {
    "U - Stanza": "#50b131",
    "U - Trankit": "#f77189",
    "N - Gold": "#bb9832",
    "N - Gold & Silver": "#464d49",
    "N - Gold (strict)": "#3ba3ec",
    "N - Gold & Silver (strict)": "#e866f4",
}

fig, axes = plt.subplots(4, 1, figsize=(10, 12), dpi=700)

for i, lang in enumerate(["de", "nl", "it", "en"]):
    df = pd.read_csv(f"{lang}_tok.csv")
    ax = sns.lineplot(
        x="n_char",
        y="value",
        hue="ud_system",
        ci=None,
        data=pd.melt(
            df[["f1", "n_char", "ud_system"]],
            ["n_char", "ud_system"],
            var_name="Scores",
        ),
        palette=palette,
        linewidth=1.0,
        ax=axes[placement[i]],
        legend=i == 3,
    )
    ax.set(ylim=(0, 1.0), xlim=(0, 230), xlabel="", ylabel="")
    if i != 3:
        ax.tick_params(bottom=False)
        ax.set(xticklabels=[])

    if i == 3:
        # ax.set(xlabel="Number of tokens")
        ax.set(xlabel="Number of characters")
        # Put a legend below current axis
        ax.legend(loc="upper center", bbox_to_anchor=(0.5, -0.25), ncol=3, facecolor="white")
        # , fontsize="10", 

    ax.set(ylabel=lang_mapping[lang])
    ax.set_ylabel(ax.get_ylabel(), rotation=0, horizontalalignment="right")

fig.supylabel("")
fig.suptitle("")
fig.tight_layout()

fig.savefig(f"analysis/scores_vs_len_char.pdf")
```
